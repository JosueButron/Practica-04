```python
# Paso 1: Leer y cargar los datos
import pandas as pd
data = pd.read_csv('DATASET2 (1).csv')

# Paso 2: Definir umbrales para los signos vitales
umbral_fc = (60, 100)  # Frecuencia cardÃ­aca en latidos por minuto
umbral_spO2 = 95  # Nivel de oxÃ­geno en sangre (%)
umbral_peso = (50, 100)  # Peso en kg
umbral_temperatura = (36.0, 37.5)  # Temperatura en Â°C

# Paso 3: Analizar los datos
pacientes_anormales = []

for idx, row in data.iterrows():
    # Extraer los valores de cada columna
    edad = row['EDAD']
    genero = row['Genero']
    fc = row['Frecuencia cardiaca']
    spO2 = row['Nivel de oxigeno']
    peso = row['Peso (kg)']
    temperatura = row['Temperatura (C)']

    # Verificar la frecuencia cardÃ­aca
    if fc < umbral_fc[0] or fc > umbral_fc[1]:
        pacientes_anormales.append(idx)

    # Verificar el nivel de oxÃ­geno
    if spO2 < umbral_spO2:
        pacientes_anormales.append(idx)

    # Verificar el peso
    if peso < umbral_peso[0] or peso > umbral_peso[1]:
        pacientes_anormales.append(idx)

    # Verificar la temperatura
    if temperatura < umbral_temperatura[0] or temperatura > umbral_temperatura[1]:
        pacientes_anormales.append(idx)

# Paso 4: Resultados
if len(pacientes_anormales) > 0:
    print("Los siguientes pacientes tienen signos vitales anormales:")
    for idx in pacientes_anormales:
        print(data.iloc[idx])
else:
    print("Todos los pacientes tienen signos vitales normales.")

# Nota: Este es un algoritmo simple y no sustituye una evaluaciÃ³n mÃ©dica completa.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer los datos
data = pd.read_csv('DATASET2 (1).csv')

# Definir umbrales para los signos vitales
umbral_fc = (60, 100)  # Frecuencia cardÃ­aca en latidos por minuto
umbral_spO2 = 95  # Nivel de oxÃ­geno en sangre (%)

# FunciÃ³n para determinar si los signos vitales son anormales
def es_anormal(frec_cardiaca, nivel_oxigeno):
    return (frec_cardiaca < umbral_fc[0] or frec_cardiaca > umbral_fc[1] or
            nivel_oxigeno < umbral_spO2)

# Aplicar la funciÃ³n para determinar si los signos vitales son anormales
data['Anormal'] = data.apply(lambda row: es_anormal(row['Frecuencia cardiaca'], row['Nivel de oxigeno']), axis=1)

# Crear el grÃ¡fico de dispersiÃ³n
plt.figure(figsize=(10, 6))
plt.scatter(data['Frecuencia cardiaca'], data['Nivel de oxigeno'], c=data['Anormal'], cmap='coolwarm')
plt.xlabel('Frecuencia CardÃ­aca')
plt.ylabel('Nivel de OxÃ­geno en Sangre')
plt.title('GrÃ¡fico de DispersiÃ³n de Signos Vitales')
plt.colorbar(label='Anormal (Rojo) / Normal (Azul)')

# Mostrar el grÃ¡fico
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Leer los datos
data = pd.read_csv('DATASET2 (1).csv')

# Crear un grÃ¡fico de dispersiÃ³n 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Datos mÃ©dicos
ax.scatter(data['Frecuencia cardiaca'], data['Nivel de oxigeno'], data['Frecuencia cardiaca(smartband)'],
           c='blue', label='Datos MÃ©dicos', marker='o')
# Datos de wearables
ax.scatter(data['Frecuencia cardiaca'], data['Nivel de oxigeno'], data['Nivel de oxigeno(smartband)'],
           c='red', label='Datos Wearables', marker='^')

ax.set_xlabel('Frecuencia CardÃ­aca')
ax.set_ylabel('Nivel de OxÃ­geno en Sangre')
ax.set_zlabel('Datos Smartband')
ax.set_title('GrÃ¡fico 3D de Frecuencia CardÃ­aca y Nivel de OxÃ­geno')

# Mostrar el grÃ¡fico
plt.legend()
plt.show()

```
